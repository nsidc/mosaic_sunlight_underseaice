# Convert Time Units in NetCDF

The MOSAiC met data is 1 minute sampling frequency.  These files are somewhat large.

```python
import numpy as np
import matplotlib.pyplot as plt

import xarray as xr

from mosaic_underice_sunlight.filepath import MET_TOWER_DATAPATH
```

```python
filelist = list(MET_TOWER_DATAPATH.glob('mosmet*.nc'))
```

```python
fsize = np.array([f.stat().st_size for f in filelist])
np.quantile(fsize, [0.1,0.5,0.9])
```

```python
%%time
ds = xr.open_dataset(filelist[20])
```

```python
ds
```

```python
%%time
ds = xr.open_mfdataset(filelist)
```

```python
%%time
fig, ax = plt.subplots(figsize=(12,7))
ds.temp_2m.plot(ax=ax, label="2m Air Temperature")
ds.skin_temp_surface.plot(ax=ax, label="Skin Temperature")
ax.axhline(0., color='0.5')
ax.legend()
```
